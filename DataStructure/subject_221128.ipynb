### 1) 원소들이 다음 순서로 리스트에 저장되어 있을때 수행되는 과정

12 70 30 20 55 25

```python

# 선택 정렬
def selectionSort(A) -> int:
    count1 = 0
    printArray(A)
    for last in range(len(A)-1,0,-1):
        k = theLargest(A,last)
        A[k],A[last] = A[last],A[k]
        printArray(A)
        count1 += 1
    return count1    
     
        

def theLargest(A,last:int) -> int:
    largest = 0
    for i in range(last+1):
        if A[i] > A[largest]:
            largest = i
    return largest  

# 버블 정렬
def bubbleSort(A) -> int:
    count2 = 0
    printArray(A)
    for numElement in range(len(A),0,-1):
        for i in range(numElement-1):
            if A[i] > A[i+1]:
                A[i],A[i+1] = A[i+1],A[i]
                printArray(A)
                count2 += 1
    return count2            
              

# 삽입 정렬
def insertSort(A) ->  int:
    count3 = 0
    printArray(A)
    for i in range(1,len(A)):
        location = i-1 # 앞의 원소
        data = A[i] # 뒤의 원소
        while location >= 0 and data < A[location]: # 뒤 원소가 앞 원소보다 작을때 까지
            A[location+1] = A[location] # 하나씩 뒤로 옮기기
            location -= 1 # 위치 앞으로 옮기기
            printArray(A)
            count3 +=1
        A[location+1] = data   # 뒤 원소가 앞 원소보다 작으면 앞 원소 자리에 뒤 원소 데이터 입력
        printArray(A)
        count3 += 1
    return count3    
       

def printArray(A):
    for data in A:
        print(data, end = " ")
    print(end = "->")        


if __name__ == "__main__":


    A = [12,70,30,20,55,25]
    B = [12,70,30,20,55,25]
    C = [12,70,30,20,55,25]

    
    print('\n선택 정렬: '+str(selectionSort(A))+' 번 정렬') # 선택 정렬
    print('\n버블 정렬: '+str(bubbleSort(B))+' 번 정렬') # 선택 정렬
    print('\n삽입정렬: '+str(insertSort(C))+' 번 정렬')# 삽입 정렬
```

### 3) 원소들이 다음 순서로 리스트에 저장되어 있을때 병합 정렬/ 힙 정렬이 수행되는 과정을 보이시오

12 70 30 20 55 25 40 50

```python
def printArray(A):
    for data in A:
        print(data, end = " ")
    print(end = "->")  

def mergeSort(A,p:int,r:int):
    if p < r:
        q = (p+r)//2
        mergeSort(A,p,q)
        printArray(A)
        mergeSort(A,q+1,r)
        printArray(A)
        merge(A,p,q,r)
        printArray(A)

def merge(A,p:int,q:int,r:int):
    i=p; j=q+1; t = 0
    tmp = [0 for i in range(len(A))]
    while i<= q and j<=r:
        if(A[i]<=A[j]): # A1의 n번째 원소가 A2의 n번째 원소보다 작으면
            tmp[t] = A[i] # A의 n번째 원소 추가
            t += 1; i += 1
            
        else:
            tmp[t] = A[j] # 아닐 경우 A2의 n번째 원소 추가
            t += 1; j+= 1
            
    while i <= q: # A1의 리스트가 남은 경우
        tmp[t] = A[i]
        t += 1; i += 1
        
    while j <= r: # A2의 리스트가 남은 경우
        tmp[t] = A[j]
        t += 1; j+= 1

    i = p; t = 0    # 초기화
    while i <= r : # 처음부터 끝까지
        A[i] = tmp[t] # A로 복사
        i+= 1; t+= 1
        
def quickSort(A,p:int,r:int):
    if p < r:
        q = partition(A,p,r) # 분할
        # q = 기준 원소의 인덱스
        printArray(A)
        quickSort(A,p,q-1) # 기준 원소 이전 부분 리스트 정렬
        quickSort(A,q+1,r) # 기준 원소 이후 부분 리스트 정렬

def partition(A,p:int,r:int) -> int:
    x = A[r] # 기준 원소        
    i = p-1 # i는 1구역의 끝지점
    
    for j in range(p,r): # 3구역 끝지점 까지
        if A[j] < x: # 아직 판단되지 않은 원소가 기준 원소보다 작으면
            i += 1 # 인덱스 추가후
            A[i],A[j] = A[j],A[i] # 원소 교환
            printArray(A)
    A[i+1],A[r] = A[r],A[i+1] # 기준원소와 2구역 첫 원소 교환 (기준원소가 자기보다 작은 원소와 큰 원소 사이에 오게끔)
    printArray(A)
    return i+1  # 기준원소의 인덱스 return 
                 

if __name__ == "__main__":
    A = [12,70,30,20,55,25,40,50]
    B = [12,70,30,20,55,25,40,50]

    print("병합 정렬의 정렬 과정: ")
    mergeSort(A,0,len(A)-1)
    print()
    print("퀵 정렬의 정렬 과정")
    quickSort(B,0,len(B)-1)

```

### 5) 교과서 281쪽과 같은 알고리즘 성능 비교를 수행하고자 한다.
### 각각 1000, 10000, 100000의 크기에 대해 실행하되 100회의 평균을 구하라

```python
from random import randrange
import random
import time
import math

def selectionSort(A):
    for last in range(len(A)-1,0,-1):
        k = theLargest(A,last)
        A[k],A[last] = A[last],A[k]

def theLargest(A,last:int) -> int:
    largest = 0
    for i in range(last+1):
        if A[i] > A[largest]:
            largest = i
    return largest   

def bubbleSort(A):
    for numElement in range(len(A),0,-1):
        for i in range(numElement-1):
            if A[i] > A[i+1]:
                A[i],A[i+1] = A[i+1],A[i]    

def insertSort(A):
    for i in range(1,len(A)):
        location = i-1 # 앞의 원소
        data = A[i] # 뒤의 원소
        while location >= 0 and data < A[location]: # 뒤 원소가 앞 원소보다 작을때 까지
            A[location+1] = A[location] # 하나씩 뒤로 옮기기
            location -= 1 # 위치 앞으로 옮기기
        A[location+1] = data   # 뒤 원소가 앞 원소보다 작으면 앞 원소 자리에 뒤 원소 데이터 입력                

def mergeSort(A,p:int,r:int):
    if p < r:
        q = (p+r)//2
        mergeSort(A,p,q)
        mergeSort(A,q+1,r)
        merge(A,p,q,r)

def merge(A,p:int,q:int,r:int):
    i=p; j=q+1; t = 0
    tmp = [0 for i in range(len(A))]
    while i<= q and j<=r:
        if(A[i]<=A[j]): # A1의 n번째 원소가 A2의 n번째 원소보다 작으면
            tmp[t] = A[i] # A의 n번째 원소 추가
            t += 1; i += 1
        else:
            tmp[t] = A[j] # 아닐 경우 A2의 n번째 원소 추가
            t += 1; j+= 1
    while i <= q: # A1의 리스트가 남은 경우
        tmp[t] = A[i]
        t += 1; i += 1
    while j <= r: # A2의 리스트가 남은 경우
        tmp[t] = A[j]
        t += 1; j+= 1

    i = p; t = 0    # 초기화
    while i <= r : # 처음부터 끝까지
        A[i] = tmp[t] # A로 복사
        i+= 1; t+= 1

def quickSort(A,p:int,r:int):
    if p < r:
        q = partition(A,p,r) # 분할
        # q = 기준 원소의 인덱스
        quickSort(A,p,q-1) # 기준 원소 이전 부분 리스트 정렬
        quickSort(A,q+1,r) # 기준 원소 이후 부분 리스트 정렬

def partition(A,p:int,r:int) -> int:
    x = A[r] # 기준 원소        
    i = p-1 # i는 1구역의 끝지점
    
    for j in range(p,r): # 3구역 끝지점 까지
        if A[j] < x: # 아직 판단되지 않은 원소가 기준 원소보다 작으면
            i += 1 # 인덱스 추가후
            A[i],A[j] = A[j],A[i] # 원소 교환
    A[i+1],A[r] = A[r],A[i+1] # 기준원소와 2구역 첫 원소 교환 (기준원소가 자기보다 작은 원소와 큰 원소 사이에 오게끔)
    return i+1  # 기준원소의 인덱스 return 

def percolateDown(A,k:int,end:int):
    child = 2*k +1
    right = 2*k +2

    if child <= end :
        if right <= end and A[child] < A[right]:
            child = right
        if A[k] < A[child]:
            A[k],A[child] = A[child],A[k]
            percolateDown(A,child,end)

def buildHeap(A):
    for i in range((len(A)-2)//2,-1,1):
        percolateDown(A,i,len(A)-1)

def heapSort(A):
    buildHeap(A)
    for last in range (len(A)-1,0,-1):
        A[last],A[0] = A[0],A[last]
        percolateDown(A,0,last-1)             


def insertSort(A):
    for i in range(1,len(A)):
        location = i-1 # 앞의 원소
        data = A[i] # 뒤의 원소
        while location >= 0 and data < A[location]: # 뒤 원소가 앞 원소보다 작을때 까지
            A[location+1] = A[location] # 하나씩 뒤로 옮기기
            location -= 1 # 위치 앞으로 옮기기
        A[location+1] = data   # 뒤 원소가 앞 원소보다 작으면 앞 원소 자리에 뒤 원소 데이터 입력 

def bucketSort(A):
    n = len(A)
    B = [[] for _ in range(n)]
    for i in range(n):
        B[math.floor(n*A[i])].append(A[i])
    A.clear()
    for i in range(n):
        insertSort(B[i])
        A.extend(B[i])    
        




for i in range(10000):
    b2.append(randrange(100))
    i2.append(randrange(100))
    m2.append(randrange(100))
    q2.append(randrange(100))
    h2.append(randrange(100))
    e.append(random.random())

for i in range(100000):
    b3.append(randrange(100))
    i3.append(randrange(100))
    m3.append(randrange(100))
    q3.append(randrange(100))
    h3.append(randrange(100))
    f.append(random.random())   


bubble_time = [0 for _ in range(3)]
insert_time = [0 for _ in range(3)]
merge_time = [0 for _ in range(3)]
quick_time = [0 for _ in range(3)]
heap_time = [0 for _ in range(3)]
bucket_time = [0 for _ in range(3)]

for j in range(100):
    for i in range(1000):
        b1 = []
        b1.append(randrange(100))
    bubble_time_start = time.time()
    bubbleSort(b1)
    bubble_time[0] += time.time() - bubble_time_start

    for i in range(1000):
        i1 = []
        i1.append(randrange(100))
    insert_time_start = time.time()
    insertSort(i1)
    insert_time[0] += time.time() - insert_time_start

    for i in range(1000):
        m1 = []
        m1.append(randrange(100))
    merge_time_start = time.time()
    mergeSort(m1,0,len(m1)-1)
    merge_time[0] += time.time() - merge_time_start

    for i in range(1000):
        q1 = []
        q1.append(randrange(100))
    quick_time_start = time.time()
    quickSort(q1,0,len(q1)-1)
    quick_time[0] += time.time() - quick_time_start

    for i in range(1000):
        h1 = []
        h1.append(randrange(100))
    heap_time_start = time.time()
    heapSort(h1)
    heap_time[0] += time.time() - heap_time_start

    for i in range(1000):
        d = []
        d.append(random.random())
    bucket_time_start = time.time()
    bucketSort(d)
    bucket_time[0] += time.time() - bucket_time_start






bubble_time[0] = bubble_time[0]/100
insert_time[0] = insert_time[0]/100
merge_time[0] = merge_time[0]/100
quick_time[0] = quick_time[0]/100
heap_time[0] = heap_time[0]/100
bucket_time[0] = bucket_time[0]/100

print('크기 1000 평균 시간: ')
print('버블정렬: ',bubble_time[0], ' 초')
print('삽입정렬: ',insert_time[0], ' 초')
print('병합정렬: ',merge_time[0], ' 초')
print('퀵정렬: ',quick_time[0], ' 초')
print('힙정렬: ',heap_time[0], ' 초')
print('버킷정렬: ',bucket_time[0], ' 초')

for i in range(100):

    for i in range(10000):
        e = []
        e.append(random.random())
    bucket_time_start = time.time()
    bucketSort(e)
    bucket_time[1] += time.time() - bucket_time_start

    for i in range(10000):
        b2 = []
        b2.append(randrange(100))
    bubble_time_start = time.time()
    bubbleSort(b2)
    bubble_time[1] += time.time() - bubble_time_start

    for i in range(10000):
        i2 = []
        i2.append(randrange(100))
    insert_time_start = time.time()
    insertSort(i2)
    insert_time[1] += time.time() - insert_time_start

    for i in range(10000):
        m2 = []
        m2.append(randrange(100))
    merge_time_start = time.time()
    mergeSort(m2,0,len(m2)-1)
    merge_time[1] += time.time() - merge_time_start

    for i in range(10000):
        q2 = []
        q2.append(randrange(100))
    quick_time_start = time.time()
    quickSort(q2,0,len(q2)-1)
    quick_time[1] += time.time() - quick_time_start

    for i in range(10000):
        h2 = []
        h2.append(randrange(100))
    heap_time_start = time.time()
    heapSort(h2)
    heap_time[1] += time.time() - heap_time_start

bubble_time[1] = bubble_time[1]/100
insert_time[1] = insert_time[1]/100
merge_time[1] = merge_time[1]/100
quick_time[1] = quick_time[1]/100
heap_time[1] = heap_time[1]/100
bucket_time[1] = bucket_time[1]/100

print('크기 10000 평균 시간: ')
print('버블정렬: ',bubble_time[1], ' 초')
print('삽입정렬: ',insert_time[1], ' 초')
print('병합정렬: ',merge_time[1], ' 초')
print('퀵정렬: ',quick_time[1], ' 초')
print('힙정렬: ',heap_time[1], ' 초')
print('버킷정렬: ',bucket_time[1], ' 초')


for i in range(100):
    for i in range(100000):
        f = []
        f.append(random.random())
    bucket_time_start = time.time()
    bucketSort(f)
    bucket_time[2] += time.time() - bucket_time_start

    for i in range(100000):
        m3 = []
        m3.append(randrange(100))
    merge_time_start = time.time()
    mergeSort(m3,0,len(m3)-1)
    merge_time[2] += time.time() - merge_time_start

    for i in range(100000):
        q3 = []
        q3.append(randrange(100))
    quick_time_start = time.time()
    quickSort(q3,0,len(q3)-1)
    quick_time[2] += time.time() - quick_time_start

    for i in range(100000):
        h3 = []
        h3.append(randrange(100))
    heap_time_start = time.time()
    heapSort(h3)
    heap_time[2] += time.time() - heap_time_start


    for i in range(100000):
        b3 = []
        b3.append(randrange(100))
    bubble_time_start = time.time()
    bubbleSort(b3)
    bubble_time[2] += time.time() - bubble_time_start

    for i in range(100000):
        i3 = []
        i3.append(randrange(100))
    insert_time_start = time.time()
    insertSort(i3)
    insert_time[2] += time.time() - insert_time_start

bubble_time[2] = bubble_time[2]/100
insert_time[2] = insert_time[2]/100
merge_time[2] = merge_time[2]/100
quick_time[2] = quick_time[2]/100
heap_time[2] = heap_time[2]/100
bucket_time[2] = bucket_time[2]/100

print('크기 100000 평균 시간: ')
print('버블정렬: ',bubble_time[2], ' 초')
print('삽입정렬: ',insert_time[2], ' 초')
print('병합정렬: ',merge_time[2], ' 초')
print('퀵정렬: ',quick_time[2], ' 초')
print('힙정렬: ',heap_time[2], ' 초')
print('버킷정렬: ',bucket_time[2], ' 초')

    

```
