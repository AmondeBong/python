## 스택 예제

### 헨젤과 그레텔의 집으로 돌아가기

```python
import random
def isStackFull():
    global SIZE,stack,top
    if top >= SIZE-1:
        return True
    else:
        return False

def isStackEmpty():
    global SIZE,stack,top        
    if top == -1:
        print("스택이 비었음")
        return True
    else :
        return False

def push(data):
    global SIZE,stack,top
    if(isStackFull()):
        return
    else:
        top += 1
        stack[top] = data
        

def pop():
    global SIZE,stack,top 
    if(isStackEmpty()):
        return
    else:
        data = stack[top]
        stack[top] = None
        top -= 1
        return data    

def peek():
    global SIZE,top,stack
    if(isStackEmpty()):
        return
    else:
        return stack[top]

color = ['빨강','주황','노랑','초록','파랑','보라']
SIZE = 6
stack = [None for _ in range(SIZE)]
top = -1

if __name__ == "__main__":

    print("과자집에 가는길 :",end = " ")
    while True:
        stage = color[random.randint(0,5)]
        if stage not in stack:
            push(stage)
            print(stage, end = " --> ")
        if (isStackFull()):
            break
        
    print("과자집",end = " ")
    print()

    print("우리집에 오는 길 :", end = " ")
    for i in range(len(stack)):
        print(pop(),end = " --> ")
    print("우리집",end = " ")
    print()            


```
